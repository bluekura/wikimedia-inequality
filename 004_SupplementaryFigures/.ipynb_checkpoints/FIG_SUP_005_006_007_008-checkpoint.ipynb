```python
%matplotlib inline  
import glob
import zipfile
#import multiprocessing as mp
import gc
#from collections import defaultdict
import csv
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
import numpy as np
```

```python
final_table = pd.read_csv("./data/Allwiki_output_modify_langinfo.txt", index_col=0)
```

```python
avg_table = (final_table.groupby('lang_code').mean() )
avg_table
```

```python
types = ['wiki', 'wiktionary', 'wikibooks', 'wikiquote', 'wikisource', 'wikinews', 'wikivoyage', 'wikiversity']
types_x = ['wiki', 'wiktionary', 'wikibooks']
types_map = {'wiki':'Wikipedia',
             'wiktionary':'Wiktionary',
             'wikibooks':'Wikibooks', 
             'wikiquote':'Wikiquote',
             'wikisource':'Wikisource',
             'wikinews':'Wikinews', 
             'wikivoyage':'Wikivoyage',
             'wikiversity':'Wikiversity'}
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(15,5))

for i in range(len(types_x)):    
    for j in range(len(types)):
        if(i==j):
            continue
        ax = axes[i]
        ax.set_xscale('log')
        ax.set_yscale('log')
        #plt.title(types[i])
        ax.set_xlabel("$N_e$ (" + types_map[types[i]] + ")")
        ax.set_ylabel("$N_e$")
        temp_table_x = final_table[:][(final_table["type"] == types_x[i]) & (final_table["lang_name"] != "")][['lang_code', '3:Num_Edit']] 
        temp_table_y = final_table[:][(final_table["type"] == types[j]) & (final_table["lang_name"] != "")][['lang_code', '3:Num_Edit']]
        temp_table = temp_table_x.merge(temp_table_y, on="lang_code")
        #temp_table.set_index('lang_code')
        #print temp_table[0:10]
        #temp_table = temp_table.div(axis = 'index')
        #print temp_table[0:10]
        ax.plot(temp_table['3:Num_Edit_x'], temp_table['3:Num_Edit_y'], "o", label = types_map[types[j]], color = 'C'+str(j), alpha=0.5)  
        ax.legend(fontsize=7)   
        
index = 0
labellist = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m"]
for ax in fig.get_axes():
    ax.get_yaxis().set_label_coords(-0.1,0.5)
    ax.yaxis.label.set_fontsize(14)
    ax.xaxis.label.set_fontsize(14)
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14)    
    for tick in ax.xaxis.get_majorticklabels():
        tick.set_horizontalalignment("right")
#    ax.legend(loc=2)
    ax.text(-0.15, 0.95, labellist[index], fontsize=14, weight='bold', transform=ax.transAxes)
    SHIFT = 0.1
    index += 1

plt.tight_layout()
plt.savefig("FIG_SUP_005.pdf")

plt.show()        
```
