```python
%matplotlib inline  
import glob
import zipfile
import numpy as np
import gc
#from collections import defaultdict
import csv
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats
#import powerlaw
```

```python
final_table = pd.read_csv("./data/Allwiki_output_modify_langinfo.txt", sep=",")
```

```python
types = ['wiki', 'wiktionary', 'wikibooks', 'wikiquote', 'wikisource', 'wikinews', 'wikivoyage', 'wikiversity']
types_index = [1,2,3,4,5,6,7,8,9]
types_name = ["All", 'Wikipedia', 'Wiktionary', 'Wikibooks', 'Wikiquote', 'Wikisource', 'Wikinews', 'Wikivoyage', 'Wikiversity']
types_map = {'wiki':'Wikipedia',
             'wiktionary':'Wiktionary',
             'wikibooks':'Wikibooks', 
             'wikiquote':'Wikiquote',
             'wikisource':'Wikisource',
             'wikinews':'Wikinews', 
             'wikivoyage':'Wikivoyage',
             'wikiversity':'Wikiversity'}
types_map_index = {'all':1,
             'wiki':2,
             'wiktionary':3,
             'wikibooks':4, 
             'wikiquote':5,
             'wikisource':6,
             'wikinews':7, 
             'wikivoyage':8,
             'wikiversity':9,}
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15,10))
outfile = "./grpout_rev2/power_fit_info_ne_np.pdf"

graph = pd.DataFrame(columns=["slope", "intercept", "rvalue", "pvalue", "stderr", "rawrvalue"])
xindex = '3:Num_Edit'
yindex = '4:Num_Editor'
axes[0,0].text(0.02, 0.92, "$N_e$-$N_p$", fontsize=20, weight='bold', transform=axes[0,0].transAxes)

graph.set_value(1, "slope", scipy.stats.linregress(np.log(final_table[xindex]), np.log(final_table[yindex])).slope)
graph.set_value(1, "intercept", np.exp(scipy.stats.linregress(np.log(final_table[xindex]), np.log(final_table[yindex])).intercept))
graph.set_value(1, "rvalue", scipy.stats.linregress(np.log(final_table[xindex]), np.log(final_table[yindex])).rvalue)
graph.set_value(1, "pvalue", scipy.stats.linregress(np.log(final_table[xindex]), np.log(final_table[yindex])).pvalue)
graph.set_value(1, "stderr", scipy.stats.linregress(np.log(final_table[xindex]), np.log(final_table[yindex])).stderr)
graph.set_value(1, "rawrvalue", scipy.stats.pearsonr(final_table[xindex], final_table[yindex])[0])
for tp in types:
    temp_table = final_table[:][(final_table["type"] == tp) & (final_table["lang_name"] != "")]
    cur_type = types_map[tp] 
    cur_type_idx = types_map_index[tp] 
    graph.set_value(cur_type_idx, "slope", scipy.stats.linregress(np.log(temp_table[xindex]), np.log(temp_table[yindex])).slope)
    graph.set_value(cur_type_idx, "intercept", np.exp(scipy.stats.linregress(np.log(temp_table[xindex]), np.log(temp_table[yindex])).intercept))
    graph.set_value(cur_type_idx, "rvalue", scipy.stats.linregress(np.log(temp_table[xindex]), np.log(temp_table[yindex])).rvalue)
    graph.set_value(cur_type_idx, "pvalue", scipy.stats.linregress(np.log(temp_table[xindex]), np.log(temp_table[yindex])).pvalue)    
    graph.set_value(cur_type_idx, "stderr", scipy.stats.linregress(np.log(temp_table[xindex]), np.log(temp_table[yindex])).stderr)    
    graph.set_value(cur_type_idx, "rawrvalue", scipy.stats.pearsonr(temp_table[xindex], temp_table[yindex])[0])

#print graph
graph.index_name = "Type"
axes[0,0].set_xticks(types_index)
axes[0,0].set_title("Exponent")
axes[0,0].set_xticklabels(types_name, rotation=40)
axes[0,0].scatter(types_index, graph.slope)

axes[0,1].scatter(types_index, graph.pvalue)
axes[0,1].set_title("p-value")
axes[0,1].set_ylim(0, 5e-07)
axes[0,1].set_xticks(types_index)
axes[0,1].set_xticklabels(types_name, rotation=40)

axes[0,2].scatter(types_index, graph.stderr)
axes[0,2].set_title("Standard error")
axes[0,2].set_xticks(types_index)
axes[0,2].set_xticklabels(types_name, rotation=40)

axes[1,0].scatter(types_index, graph.intercept)
axes[1,0].set_title("Proportionality coefficient")
axes[1,0].set_xticks(types_index)
axes[1,0].set_xticklabels(types_name, rotation=40)

axes[1,1].scatter(types_index, graph.rawrvalue)
axes[1,1].set_title("Pearson correlation r")
axes[1,1].set_xticks(types_index)
axes[1,1].set_xticklabels(types_name, rotation=40)

axes[1,2].scatter(types_index, graph.rvalue)
axes[1,2].set_title("Pearson correlation r (log-log)")
axes[1,2].set_xticks(types_index)
axes[1,2].set_xticklabels(types_name, rotation=40)


plt.tight_layout()
plt.savefig(outfile)
plt.show()

print graph
```
